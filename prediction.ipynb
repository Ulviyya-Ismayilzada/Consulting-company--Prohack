```python
X=train.iloc[:,0:8]
z=train['y']
```

```python
 from sklearn.model_selection import train_test_split
X_train,X_test,z_train,z_test=train_test_split(X,z,test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import mutual_info_regression
best_features = SelectKBest(score_func=mutual_info_regression, k=5)
X_train= best_features.fit_transform(X_train, z_train)
```

```python
X_test= best_features.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
model.fit(X_train,z_train)
```

```python
print(model.intercept_)
```

```python
print(model.coef_)
```

```python
z_pred=model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error
```

```python
print("MAE:", round(mean_absolute_error(z_test,z_pred),2))
print("RMSE:",round(np.sqrt(mean_squared_error(z_test,z_pred)),2))
print("R2:",round(r2_score(z_test,z_pred),2))
```

```python
from sklearn.ensemble import RandomForestRegressor
model2=RandomForestRegressor()
```

```python
model2.fit(X_train,z_train)
```

```python
z_pred=model2.predict(X_test)
```

```python
print("MAE:", round(mean_absolute_error(z_test,z_pred),2))
print("RMSE:",round(np.sqrt(mean_squared_error(z_test,z_pred)),2))
print("R2:",round(r2_score(z_test,z_pred),2))
```
